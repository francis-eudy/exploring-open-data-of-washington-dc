```python
import json
import pandas as pd
import numpy as np
import urllib3
from urllib3 import request
import certifi
```

```python
# Moving Violations Issued in March 2021
url = 'https://maps2.dcgis.dc.gov/dcgis/rest/services/DCGIS_DATA/Violations_Moving_2021/MapServer/2/query?where=1%3D1&outFields=*&outSR=4326&f=json'
```

```python
http = urllib3.PoolManager(
       cert_reqs='CERT_REQUIRED',
       ca_certs=certifi.where())
```

```python
r = http.request('GET', url)
r.status
```

```python
data = json.loads(r.data.decode('utf-8'))
data
```

```python
df = pd.json_normalize(data, 'features')
df.head(10)
```

```python
df.columns
```

```python
df.columns = df.columns.str.lstrip('attributes.')
df.columns
```
